```python
f_array = []
fw_array = []
j= []
p = []
clean_array = []
y =0

with open('headlines.txt') as my_file:
    f_array = my_file.read().split('\n')
    my_file.close() 

for i in f_array:
    j = i
    x = j.split()
    fw_array.append(x)

clean_array = fw_array

for k in range(len(fw_array)):
    for l in range(len(fw_array[k])):
        #print(fw_array[k][l])
        az = r.sub(r'[^a-zA-Z ]', '',fw_array[k][l])
        clean_array[k][l] = az

#print(clean_array)



```
